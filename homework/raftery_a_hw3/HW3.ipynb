#### Problem 1(f)

Build a simple linear regression model to predict the number of wins from the average adjusted singles, double, triples, HR, and BB rates. To decide which of these terms to include fit the model to data from 2002 and compute the average squared residuals from predictions to years past 2012. Use the fitted model to define a new sabermetric summary: offensive predicted wins (OPW). Hint: the new summary should be a linear combination of one to five of the five rates.


```python
### Your code here ###
import statsmodels.api as sm
import itertools

normalized_stats = normalized_stats.reset_index()
stats1 = normalized_stats[normalized_stats['yearID'] < 2002]
stats2 = normalized_stats[normalized_stats['yearID'] >= 2002]


Y = stats1['W']
regressors = ['BB', '1B', '2B', '3B', 'HR']

# make a generator of all possible combinations of the regressors
# e.g. [[BB], [BB, 1B], [BB, 1B, 2B]...etc.]
combinations = itertools.chain(*[itertools.combinations(regressors, i) for i in range(1, len(regressors) + 1)])
models = []

# make a regression model for each combination of regressors
for regs in combinations:
    X = stats1[list(regs)]
    X = sm.add_constant(X)
    model = sm.OLS(Y,X).fit()
    models.append(model)

# calculate the MSE by applying the models
# to the testing data, and choose the best model
best_model = 0
mses = {}
for i, model in enumerate(models):
    regs = model.params.index[1:]
    X = stats2[regs]
    X = sm.add_constant(X)
    predictions = model.predict(X)
    errors = (predictions - stats2['W']) ** 2
    mse = errors.mean()
    if not mses or mse < np.min(mses.values()):
        best_model = i
    mses[i] = mse

best_model = models[best_model]
best_model.summary()
```

```python
def to_equation(model, yvar):
    return '{} = '.format(yvar) + ' + '.join(['({} * {})'.format(*i) for i in zip(model.params[::-1][:-1], model.params.index[::-1][:-1])]) + ' + {}'.format(model.params[0])

X = stats2[best_model.params.index[1:]]
X = sm.add_constant(X)
predictions = best_model.predict(X)
actual = stats2['W']

print to_equation(best_model, 'W')

plt.scatter(predictions, actual)
plt.title('Actual vs. Predicted Wins Using Best Model, 2002 - 2012 Data')
plt.xlabel('Actual Wins')
plt.ylabel('Predicted Wins')
plt.xlim([55, 105])
plt.ylim([55, 105])
plt.show()
```

```python
# calcualte "OPW" for all stats
X = normalized_stats[best_model.params.index[1:]]
X = sm.add_constant(X)
predictions = best_model.predict(X)
normalized_stats['OPW'] = predictions
normalized_stats.head()
```

#### Problem 1(l)
Pick one players from one of each of these 10 position C, 1B, 2B, 3B, SS, LF, CF, RF, or OF keeping the total median salary of all 10 players below 20 million. Report their averaged predicted wins and total salary.

```python
### Your code here ###
def df_to_tuples(df):
    return [tuple(x) for x in df.reset_index()[['playerID', 'salary', 'OPW']].values]

positions = ['C', '1B', '2B', '3B', 'SS', 'LF', 'CF', 'RF', 'OF', 'DH']
position_players = {}
for position in positions:
    ps = playerLS[playerLS['POS'] == position]
    position_players[position] = df_to_tuples(ps)
```

```python
# construct team
import random
i = 0
best_opw = 0
best_team = None
def draw():
    team = []
    salaries = []
    opws = []
    for position in positions:
        player = random.choice(position_players[position])
        team.append(player[0])
        salaries.append(player[1])
        opws.append(player[2])
        if np.sum(salaries) > 20000000:
            break
    else:
        return team, opws
    return False, False

while True:
    team, opws = draw()
    if team:
        mean_opw = np.mean(opws)
        if mean_opw > best_opw:
            best_team = team
            best_opw = mean_opw
    i += 1
    if i % 1000000 == 0:
        print i
```

```python
df = playerLS.reset_index()
chosen_team = df[df['playerID'].isin(best_team)]
print """
Chosen team total salary: {}
Chosen team mean OPW: {}""".format(chosen_team['salary'].sum(), chosen_team['OPW'].mean())
display(chosen_team)
```

#### Problem 2(c) 

In the lecture we discussed how to use cross validation to estimate the optimal value for $k$ (the number of nearest neighbors to base the classification on). Use ***ten fold cross validation*** to estimate the optimal value for $k$ for the iris data set. 

**Note**: For your convenience sklearn does not only include the [KNN classifier](http://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html), but also a [grid search function](http://scikit-learn.org/stable/modules/generated/sklearn.grid_search.GridSearchCV.html#sklearn.grid_search.GridSearchCV). The function is called grid search, because if you have to optimize more than one parameter, it is common practice to define a range of possible values for each parameter. An exhaustive search then runs over the complete grid defined by all the possible parameter combinations. This can get very computation heavy, but luckily our KNN classifier only requires tuning of a single parameter for this problem set. 

```python
### Your code here ###
from sklearn.grid_search import GridSearchCV

classifier = sklearn.neighbors.KNeighborsClassifier()
parameters = {'n_neighbors': np.arange(1, 10)}
gridsearch = GridSearchCV(classifier, parameters, cv=10)
gridsearch.fit(x_train, y_train)
print 'Chosen value of K: {}'.format(gridsearch.best_params_['n_neighbors'])
```

#### Problem 2(d)

Visualize the result by plotting the score results verus values for $k$.

**This is a boxplot showing the results of cross validation for several values of K. As one can see, K = 3, which is the value chosen by the grid search, is the only K with a mean score of nearly 1.00, and a very small variance.**

Verify that the grid search has indeed chosen the right parameter value for $k$.

```python
### Your code here ###
gridsearch.grid_scores_
```

**As one can see from the data above, the grid search has chosen K = 3, the value which has both the highest mean score and the lowest standard deviation.**

#### Problem 2(e)

Test the performance of our tuned KNN classifier on the test set.

```python
### Your code here ###
best_estimator = gridsearch.best_estimator_
score = (y_test == best_estimator.predict(x_test)).mean()
print score
```

**The chosen KNN classifier scored .96 when predicting the classification of the test data. This means that it classified correctly in about 96% of the cases.**

Write a **ten-fold cross validation** to estimate the optimal value for $k$ for the digits data set. *However*, this time we are interested in the influence of the number of dimensions we project the data down as well. 

Extend the cross validation as done for the iris data set, to optimize $k$ for different dimensional projections of the data. Create a boxplot showing test scores for the optimal $k$ for each $d$-dimensional subspace with $d$ ranging from one to ten. The plot should have the scores on the y-axis and the different dimensions $d$ on the x-axis. You can use your favorite plot function for the boxplots. [Seaborn](http://web.stanford.edu/~mwaskom/software/seaborn/index.html) is worth having a look at though. It is a great library for statistical visualization and of course also comes with a [`boxplot`](http://web.stanford.edu/~mwaskom/software/seaborn/generated/seaborn.boxplot.html) function that has simple means for changing the labels on the x-axis.

```python
### Your cross validation and evaluation code here ###
from sklearn.grid_search import GridSearchCV
from sklearn.decomposition import TruncatedSVD

test_results = []

max_k = 10 # max value of k to test
max_d = 10 # max dimension to test
folds = 10

for dimensions in range(1, max_d + 1):
    
    # find the best estimator for this dimension
    svd = TruncatedSVD(n_components=dimensions)
    X_proj = svd.fit_transform(x_train)

    classifier = sklearn.neighbors.KNeighborsClassifier()
    parameters = {'n_neighbors': np.arange(1, max_k + 1)}
    gridsearch = GridSearchCV(classifier, parameters, cv=folds)
    gridsearch.fit(X_proj, y_train)
    
    best_estimator = gridsearch.best_estimator_
    optimal_k = gridsearch.best_params_['n_neighbors']
    test_results.append({
        'Dimensions': dimensions,
        'Optimal K': optimal_k,
        'Score': gridsearch.best_score_,
        'Grid Scores': gridsearch.grid_scores_[optimal_k - 1].cv_validation_scores,
    })

test_results = pd.DataFrame(test_results)
test_results
```

```python
### Your boxplot code here ###
sns.boxplot(test_results['Grid Scores'], names=test_results['Dimensions'])
plt.ylabel('Score')
plt.title('Distr. of Scores Over Several Dimensions, Using Optimal K in Each Dimension')
```
