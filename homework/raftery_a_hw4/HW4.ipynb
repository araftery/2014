#### Problem 2(b)

Compute the square root of these values in Problem 2(b) for the republican candidates in each race. Then, compute the standard deviations of the observed poll results for each race. 

```python
### Your code here ###
for slug in polls:
    data = polls[slug]
    republican = candidates.ix[slug, :]['R']
    data['republican_percentage'] = data['choices'].apply(lambda x: float(x[republican.split(' ')[-1]]) / 100.0)

theoretical_stds = {}
observed_stds = {}
expected_republican = {}
for slug in polls:
    data = polls[slug]
    p = data['republican_percentage'].mean()
    data['numobs'] = data['numobs'].astype(float)
    data['var'] = (1 / data['numobs']) * p * (1 - p)
    theoretical_stds[slug] = np.sqrt(data['var'].mean())
    observed_stds[slug] = (data['republican_percentage']).std()
    expected_republican[slug] = p
        
```

# Problem 3: Prediction and Posterior Probabilities

Perform a Bayesian analysis to predict the probability of Republicans winning in each state then provide a posterior distribution of the number of republicans in the senate.

#### Problem 3(a)

First, we define a Bayesian model for each race. The prior for the difference $\theta$ between republicans and democtrats will be $N(\mu,\tau^2)$. Say before seeing poll data you have no idea who is going to win, what should $\mu$ be? How about $\tau$, should it be large or small? 

** Your answer here: ** $\mu$ should be 0, since, before seeing any data, we must assume that each candidate has an equal chance of winning. $\tau$ should be large. Since we don't know anything about data, to be cautious we should assume a large variance.

#### Problem 3(b)

What is the distribution of $d$ conditioned on $\theta$. What is the posterior distribution of $\theta | d$?

**Hint**: Use normal approximation. 

**Your answer here:** $d|\theta = N(\theta, \sigma^2)$, where $\sigma$ is given by the SD of poll data across all races.

Using normal approxmiation, the posterior distribution is normal with mean $E(\theta|d) = B\mu + (1 - B)d$ and variance $B = \frac{\sigma^2}{\sigma^2 + \tau^2}$. $\sigma^2$ is as defined above (variance across polls from a given race), and $\mu$ and $\tau^2$ are defined as in 3(a) - 0, and a reasonably large number respectively. $\tau^2$ can be estimated by the variance across all polls.

#### Problem 3(c)

The prior represents what we think before hand. We do not know who is expected to win, so we assume $\mu=0$. For this problem estimate $\tau$ using the observed differences across states (Hint: $\tau$ represents the standard deviation of a typical difference). Compute the posterior mean for each state and plot it against original average. Is there much change? Why or why not? 

```python
### Your code here ###
ds = []
for slug in polls:
    data = polls[slug]
    data['d'] = 2 * data['republican_percentage'] - 1
    ds.append(data['d'])

ds = np.array(pd.concat(ds))
tau = ds.std()
print 'tau: {}'.format(tau)

mu = 0

posterior_means = []
for slug in polls:
    data = polls[slug]
    sigma = data['d'].std()
    B = (sigma ** 2)/((sigma ** 2) + (tau ** 2))
    d = data['d'].mean()
    posterior_mean = B * mu + (1 - B) * d
    posterior_means.append({
        'Slug': slug,
        'State': slug_to_state(slug),
        'Posterior Mean': posterior_mean,
        'sigma': sigma,
        'tau': tau,
        'B': B,
        'd': d,
    })

posterior_means = pd.DataFrame(posterior_means).sort('State')
display(posterior_means)
plt.scatter(x=posterior_means['d'], y=posterior_means['Posterior Mean'])
plt.title('Observed `d` vs. Posterior Mean for Each Race')
plt.xlabel('Observed `d`')
plt.ylabel('Posterior Mean')
plt.plot([-.6, .6], [-.4, .3], ls='--')
plt.show()
```

**Your answer here:** The posterior means are very similar to the observed values, though the posterior means seem to diverge as the observed mean differences increase. The two values are so similar because B is generally low, as one can see in the dataframe printed above. This means that $(1 - B)$ will be high, so $E(\theta|d)$ will be very similar to $d$.

#### Problem 3(d)

For each state, report a probabilty of Republicans winning. How does your answer here compare to the other aggregators?

```python
### Your code here ###
from scipy.stats import norm

posterior_means['Z Score'] = (0 - posterior_means['Posterior Mean']) / (posterior_means['B'].apply(np.sqrt))
posterior_means['Republican Probability'] = 1 - posterior_means['Z Score'].apply(norm.cdf)
display(posterior_means[['State', 'Slug', 'Republican Probability', 'Z Score', 'B', 'Posterior Mean', 'd']].sort('State'))
```

**Your answer here:** These estimates are very different from many forecasts. I think this may be because of the way I calculated $d$, awarding all undecided percentages to the Democratic candidate, thus inflating their numbers. I've computed corrected probabilities below.

```python
new_posterior_means = []

new_ds = np.array(pd.concat([polls[i]['d'] for i in polls]))
new_tau = new_ds.std()
print 'new_tau: {}'.format(new_tau)

mu = 0

for slug in polls:
    data = polls[slug]
    democrat = candidates.ix[slug, :]['D']
    data['democratic_percentage'] = data['choices'].apply(lambda x: float(x[democrat.split(' ')[-1]]) / 100.0)
    data['undecided_percentage'] = 1 - (data['democratic_percentage'] + data['republican_percentage'])
    data['d'] = data['republican_percentage'] - data['democratic_percentage']
    sigma = data['d'].std()
    B = (sigma ** 2)/((sigma ** 2) + (tau ** 2))
    d = data['d'].mean()
    posterior_mean = B * mu + (1 - B) * d
    new_posterior_means.append({
        'Slug': slug,
        'State': slug_to_state(slug),
        'Posterior Mean': posterior_mean,
        'sigma': sigma,
        'tau': new_tau,
        'B': B,
        'd': d,
    })

new_posterior_means = pd.DataFrame(new_posterior_means).sort('State')
new_posterior_means['Z Score'] = (0 - new_posterior_means['Posterior Mean']) / (new_posterior_means['B'].apply(np.sqrt))
new_posterior_means['Republican Probability'] = 1 - new_posterior_means['Z Score'].apply(norm.cdf)
display(new_posterior_means[['State', 'Slug', 'Republican Probability', 'Z Score', 'B', 'Posterior Mean', 'd']].sort('State'))
```

```python
import math
num_republicans = 30
restricted_polls = {}

def get_pollster_rank(pollster):
    try:
        return float(ranks.ix[pollster])
    except:
        return float('nan')

for slug in polls:
    data = polls[slug]
    data['rank'] = data['pollster'].apply(get_pollster_rank)
    polls_to_keep = math.ceil(.65 * len(data)) - 1
    restricted_polls[slug] = data.sort('rank').head(polls_to_keep)
```

```python
# double count polls in the last week
two_weeks_ago = dt.datetime.now() - dt.timedelta(days=7)
new_restricted_polls = {}
for slug in restricted_polls:
    data = restricted_polls[slug]
    data['enddate'] = pd.to_datetime(data['enddate'])
    new_restricted_polls[slug] = pd.concat([data, data[data['enddate'] > two_weeks_ago]])
```

```python
recent_posterior_means = []

recent_ds = np.array(pd.concat([new_restricted_polls[i]['d'] for i in new_restricted_polls]))
recent_tau = recent_ds.std()

mu = 0

for slug in new_restricted_polls:
    data = new_restricted_polls[slug]
    democrat = candidates.ix[slug, :]['D']
    data['democratic_percentage'] = data['choices'].apply(lambda x: float(x[democrat.split(' ')[-1]]) / 100.0)
    data['undecided_percentage'] = 1 - (data['democratic_percentage'] + data['republican_percentage'])
    data['d'] = data['republican_percentage'] - data['democratic_percentage']
    sigma = data['d'].std()
    B = (sigma ** 2)/((sigma ** 2) + (tau ** 2))
    d = data['d'].mean()
    posterior_mean = B * mu + (1 - B) * d
    recent_posterior_means.append({
        'Slug': slug,
        'State': slug_to_state(slug),
        'Posterior Mean': posterior_mean,
        'sigma': sigma,
        'tau': recent_tau,
        'B': B,
        'd': d,
    })

recent_posterior_means = pd.DataFrame(recent_posterior_means).sort('State')
recent_posterior_means['Z Score'] = (0 - recent_posterior_means['Posterior Mean']) / (recent_posterior_means['B'].apply(np.sqrt))
recent_posterior_means['Republican Probability'] = 1 - recent_posterior_means['Z Score'].apply(norm.cdf)
```

```python
probs = recent_posterior_means['Republican Probability'].tolist()
simulation_results = []
for _ in xrange(10000):
    republican_seats = 30
    for prob in probs:
        republican_seats += np.random.choice([1, 0], p=[prob, 1 - prob])
    simulation_results.append(republican_seats)
```
